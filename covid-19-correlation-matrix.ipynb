```python
both_data = pd.read_csv("/kaggle/input/covid19final-dataset/both_test_balanced.csv")
pcr = pd.read_csv("/kaggle/input/covid19final-dataset/pcr_balanced.csv")
rapid= pd.read_csv("/kaggle/input/covid19final-dataset/rapid_balanced.csv")

both_data_unb = pd.read_csv("/kaggle/input/covid19final-dataset/both_test_unbalanced.csv")
pcr_unb = pd.read_csv("/kaggle/input/covid19final-dataset/pcr_unbalanced.csv")
rapid_unb= pd.read_csv("/kaggle/input/covid19final-dataset/rapid_unbalanced.csv")
```

```python
def rename_columns_index(data):
    data.columns=['Sore Throat', 'Dyspnea', 'Fever',
       'Cough', 'Headache', 'Taste Disorder',
       'Olfactory Disorders', 'Coryza', 'Gender',
       'Health Professional', 'Class']

    data.index=['Sore Throat', 'Dyspnea', 'Fever',
       'Cough', 'Headache', 'Taste Disorder',
       'Olfactory Disorders', 'Coryza', 'Gender',
       'Health Professional', 'Class']
    return data
```

```python
cor_both_data=both_data.corr()
cor_both_data=cor_both_data.round(2)
cor_both_data=rename_columns_index(cor_both_data)

cor_pcr=pcr.corr()
cor_pcr=cor_pcr.round(2)
cor_pcr=rename_columns_index(cor_pcr)

cor_rapid=rapid.corr()
cor_rapid=cor_rapid.round(2)
cor_rapid=rename_columns_index(cor_rapid)

cor_both_data_unb=both_data_unb.corr()
cor_both_data_unb=cor_both_data_unb.round(2)
cor_both_data_unb=rename_columns_index(cor_both_data_unb)

cor_pcr_unb=pcr_unb.corr()
cor_pcr_unb=cor_pcr_unb.round(2)
cor_pcr_unb=rename_columns_index(cor_pcr_unb)

cor_rapid_unb=rapid_unb.corr()
cor_rapid_unb=cor_rapid_unb.round(2)
cor_rapid_unb=rename_columns_index(cor_rapid_unb)
```

```python
fig, ax =plt.subplots(1,3,figsize=(15,5))
#plt.subplots_adjust(hspace = 0.15)
sns.heatmap(cor_both_data, annot=False,linewidths=.5, cmap="YlGnBu", ax=ax[0])
ax[0].set_title('Correlation matrix of both balanced tests')
sns.heatmap(cor_pcr, annot=False,linewidths=.5,cmap="YlGnBu", ax=ax[1])
ax[1].set_title('Correlation matrix of pcr balanced test')
sns.heatmap(cor_rapid, annot=False,linewidths=.5, cmap="YlGnBu", ax=ax[2])
ax[2].set_title('Correlation matrix of rapid balanced test')
fig.tight_layout(pad=0.5)
plt.savefig('Correlation_matrix_balanced.png')
fig.show()
```

```python
fig, ax =plt.subplots(1,3,figsize=(15,5))
#plt.subplots_adjust(hspace = 0.15)
sns.heatmap(cor_both_data_unb, annot=False,linewidths=.5, ax=ax[0])
ax[0].set_title('Correlation matrix of both unbalanced tests')
sns.heatmap(cor_pcr_unb, annot=False,linewidths=.5, ax=ax[1])
ax[1].set_title('Correlation matrix of pcr unbalanced test')
sns.heatmap(cor_rapid_unb, annot=False,linewidths=.5, ax=ax[2])
ax[2].set_title('Correlation matrix of rapid unbalanced test')
fig.tight_layout(pad=0.5)
plt.savefig('Correlation_matrix_unbalanced.png')
fig.show()
```
